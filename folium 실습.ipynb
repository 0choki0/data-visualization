```python
import pandas as pd
state_unemployment='02. folium_US_Unemployment_Oct2012.csv'
state_unem_data=pd.read_csv(state_unemployment)
state_unem_data.head()
```

```python
state_geo='02. folium_us-states.json'
state_geo
```

```python
map = folium.Map(location=[40,-98], zoom_start=3)
folium.Choropleth(geo_data=state_geo,
                 data=state_unem_data,
                 columns=['State','Unemployment'],
                 key_on='feature.id',
                 fill_color='YlGn',
                 legend_name='Unemployment Rate(%)')
folium.LayerControl().add_to(map)
map
```
